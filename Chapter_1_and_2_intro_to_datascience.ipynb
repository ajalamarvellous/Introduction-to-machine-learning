```python
#opening the data file
data= open("Documents/programming/data/adult.data","r")
```

```python
#reading the content of the data object now
line1= data.readline()
print(line1)
```

```python
#reading the content into a list of lists
adult=[]                    #creating an empty list to contain lists fo each line
for x in data:              #iterating line by line of the entire data
    line=[]                 #empty list to contain each line 
    for i in x.split(","):  #splitting the content in each line, separated by commas each
        x= i.strip()        # removing the empty spaces infront and at the back of each entry
        if x.isdigit():     # booolean to check if the entry is a number
            x= int(x)       # if true, convert to an integer
        elif x=='':
            x= 0            # if empty, replace with 0
        line.append(x)      # add each of the entries to the list for each line
    adult.append(line)      # add each list of a line to the list containing the whole data
```
